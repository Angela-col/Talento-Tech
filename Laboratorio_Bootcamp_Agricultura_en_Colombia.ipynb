##Eliminación de información incongruente

###Luego del análisis de los dataframe en excel se ha detectado que se encuentran productos con producción en cero pero rendimiento positivo, por lo cual se ha hecho un análisis de impacto y al eliminar estos datos no se va a ver afectada de manera significativa.

```python
#identificación de los valores que aparecen en nulo, primer data frame
df1.isnull().sum()
```

###Para el data frame 1, en el anterior paso se identifico que en la columna "Rendimiento" se tienen 3.233 filas de datos con valores en blanco, esto en proporción con las 95.134 corresponde al 3,39% del valor total del data frame, mismos que serán eliminados para, que al momento de la visualización no generen desviaciones significativas en los análisis.**

```python
# Identificar los índices de los registros a eliminar-primer dataset
indices_eliminar = df1[(df1['Producción'] <=0)].index

# Eliminar los registros
df1 = df1.drop(indices_eliminar)

# Verificar que se hayan eliminado los registros
print(df1['Producción'].value_counts())
plt.show()
```
